# PyTorch

## What is PyTorch?
Pytorch is an open source Python-based library for implementing and running machine learning algo including wrappers for deep learning. It was primarily developed by the researchers and developers from Facebook AI Research (FAIR) lab.

PyTorch allows execution on CPUs, GPUs and XLA devices such as TPUs. It performes great when using GPUs and XLA devices. It is based on Torch library.

Pytorch has three core components:

1. Tensor library - It extends the concept of the NumPy library with the features to accelerate computation on GPUs and XLA devices. Can easily switch from CPU to GPU.

2. Automatic differentiation engine - Known as autograd, enabeling the automatic computation of gradients for tensor operations, simplifying backpropogation and model optimization. 

3. Deep learning library - It provide the building blocks including pretrained models, loss functions and optimizers for designing and training a deep learning models.

### Autograd - Automatic Differentiation

PyTorch's autograd provides functions to automatically compute gradients in dynamic computational graph.

A computational graph is a direct graph that allow us to express and visualise mathematical expressions. In deep learning, it lays outs the sequence of calculations needed to compute the output of a neural network which we will use to compute gradietns for backpropogation.

Let's look at an example to understand the concept of computational graph. We will be implementing the forward propogation of a simple logistic regression classifier, which can be seen as a single layer neural-network which returns a score between 0 and 1, which is then compared to the true class label (0 ot 1) when computing the loss.

````python
#True label
y = torch.tensor([1.0])
# Input Feature
x1 = torch.tensor([1.1])
# Weight Parameter
w1 = torch.tensor([2.2])
# Bias 
b = torch.tensor([0.0])
# Net Input
z = x1 * w1 + b
# activation and output
a = torch.sigmoid(z)
# calculate loss
loss = F.binary_cross_entropy(a,y)
print(loss)
````

When we carry out computations in PyTorch, autograd engine builds a computation graph internally if any of its terminal node has the `required_grad` attribute as True by traking every operation performed on tensors. It is useful when we have to calculate gradients which are required while training neural network by calling `grad` function.



````python
from torch.autograd import grad

#True label
y = torch.tensor([1.0])
# Input Feature
x1 = torch.tensor([1.1])
# Weight Parameter
w1 = torch.tensor([2.2], requires_grad=True)
# Bias 
b = torch.tensor([0.0], requires_grad=True)
# Net Input
z = x1 * w1 + b
# activation and output
a = torch.sigmoid(z)
# calculate loss
loss = F.binary_cross_entropy(a,y)


grad_L_w1 = grad(loss, w1, retain_graph=True)
grad_L_b = grad(loss, b, retain_graph=True)

print(grad_L_w1)
print(grad_L_b)
````

Here we have calculated grad function manually which is useful for experimentation, debugging and demonstration concepts. But PyTorch provides .backward(), a high-level tool to automate this process. We can call .backward() on loss and it will compute the gradients of all the leaf nodes in the graph which will be stored via the tensor.grad attribute.

````python
loss.backward()
print(w1.grad)
print(b.grad)
````

## Section 3: Data Handling

### Building Input Pipelines

When we are training a deep neural network, we usually train the model incrementally using an iterative optimization algo such as stochastic gradient descent. When the dataset is to large to fit into the compute memory we load the data from the main storage in chunks that is batch by batch.

Then we need to construct a data-processing pipeline to apply certain transformation and preprocessing techniques to our data, such as normalization, scaling, encoding or adding noise to augment the training procedures and to prevent overfitting.

PyTorch provides `Dataset` and `DataLoader` for efficient data handling.

### Creating Dataloader from existing tensor

If our data already exists as a tensor object, a python list or a NumPy array, we can easily create a dataset loader using `torch.utils.data.DataLoader()` class.

````python
from torch.utils.data import DataLoader

t = torch.arange(6, dtype=torch.float32)
data_loader = DataLoader(t)

for i in data_loader:
    print(i)
````

If we want to create batches of desired batch size we can do that with `batch_size` argument. The optional `drop_last` argument is useful when the elements in the tensor is not divisible by the desired batch size.

````python
t_new = torch.arange(7, dtype=torch.float32)
data_loader = DataLoader(t_new, batch_size=3, drop_last=False)
print("With batch_size = 3 and drop_last=False: ")
for i in data_loader:
    print(i)
    
data_loader = DataLoader(t_new, batch_size=3, drop_last=True)
print("\nWith batch_size = 3 and drop_last=True: ")
for i in data_loader:
    print(i)
````

### Combining two tensor into a joint dataset
Many a times we have to combine the data available in two tensors. For ex, we could have features in one tensor and lables in another. In such case, we would have to combine the tensors to build a common dataset which will allow us to retrieve the elements of these tensors in tuples.

To create a joint dataset from two tensors we need to create a `Dataset` class. A custom `Dataset` class must contain following methods:

__init__() : Constructor to initilize the internal logic, like reading array, loading files, filetring data etc.

__getitem__() : Returns the corresponding sample to the given index

````python
## We have two tensors x and y which holds features (size 3) and lables respectively.
from torch.utils.data import Dataset

x = torch.rand([4,3], dtype=torch.float32)
y = torch.arange(4)

class JointDataset(Dataset):
    
    def __init__(self, x, y):
        self.x = x
        self.y = y
    
    def __len__(self):
        return len(self.x)
    
    def __getitem__(self, idx):
        return self.x[idx], self.y[idx]

## creating an object of Jointdataset
jd_obj = JointDataset(x,y)
for img, lab in jd_obj:
    print(img, lab)
        
````

We can directly create a joint dataset by calling `TensorDataset()` directly, if one of the tensor is a labeled dataset.

````python
from torch.utils.data import TensorDataset

joint_dataset = TensorDataset(x,y)

for i in joint_dataset:
    print(i)
````

### Shuffle, Batch and Repeat

While training our NN model it is important to feed training data as randomly shuffled batches. Therefore we will shuffle and retiterate over datasets. The rows will be shuffled without loosing the one-to-one correspondense between the entire in x and y.

````python
data_loader = DataLoader(dataset= joint_dataset, batch_size=2, shuffle=True)

for i,batch in enumerate(data_loader, 1):
    print(f"Batch {i}: \nx: {batch[0]} \ny: {batch[1]}\n")
````

When we train our model for multiple epochs, we need to shuffle and iterate over the dataset by the desired number of epochs. 

````python
for epoch in range(2): 
    print(f"\nepoch {epoch+1}")
    for i, batch in enumerate(data_loader,1):
        print(f"Batch {i}: \nx: {batch[0]} \ny: {batch[1]}\n")
````

### Creating a dataset from local files

We will be creating a dataset from the images of dogs and cats we have in our local system. We will use `Image` in `PIL` to read the image file contents and transforms in `torchvision` to decode the raw contents and resize the images.

````python
# Generating a list of image file
import pathlib
## setting directory path
imgdir_path = pathlib.Path('cat_dog_images')
## iterating and yielding any jpg file in the folder
file_list = sorted([str(path) for path in imgdir_path.glob('*.jpg')])
print(file_list)
````

````python
# Vizualize images
import os
from PIL import Image

fig = plt.figure(figsize=(10,5))
for i, file in enumerate(file_list):
    img = Image.open(file)
    ax = fig.add_subplot(2, 3, i+1)
    ax.imshow(img)
    ax.set_title((os.path.basename(file), np.array(img).shape), size = 15)
    ax.axis('off')

plt.tight_layout()
plt.show()
````

We can see that these images are of different aspect ratio. Also we can see that the labels of these images are given in their filename, which we will use to assign labels to each image

````python
# labels: 1 for dog, 0 for cat
labels = [1 if 'dog' in os.path.basename(file) else 0 for file in file_list]
print(labels)
````

Now we have two lists, list of filename (path of each image) and list of their labels. We will now transform the images to be of same aspect ratio (size 80 x 120) and combine them to build single dataset.

````python
for i in file_list:
    print(Image.open(i))
````

````python
import torchvision.transforms as transforms

class CatDogImage(Dataset):
    
    def __init__(self, file_list, labels, transform = None):
        self.file_list = file_list
        self.labels = labels
        self.transform = transform
        
    def __getitem__(self, index):
        img = Image.open(self.file_list[index])
        if self.transform is not None:
            img = self.transform(img)
        label = self.labels[index]
        return img, label
    
    def __len__(self):
        return len(self.labels)
    
img_height, img_width = 80, 120

transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Resize((img_height, img_width))
])
    
image_dataset = CatDogImage(file_list, labels, transform)
````

````python
## Plotting the transformed images
fig = plt.figure(figsize=(10,6))

for i, img in enumerate(image_dataset):
    ax = fig.add_subplot(2,3,i+1)
    ax.imshow(img[0].numpy().transpose((1,2,0)))
    ax.set_title(f'{img[1]}', size = 15)
    ax.axis('off')

plt.tight_layout()
plt.savefig('cat_dog_images/08_03.pdf')
plt.show()
````
